```python
import requests, json
def get_teams_ids_for_league(league,season):

	url = "https://api-football-v1.p.rapidapi.com/v3/teams"

	querystring = {"league": league, "season":season}

	headers = {
		"X-RapidAPI-Key": "630322333bmsh894634ec4d1b679p17d2b3jsnf658c1fd7f38",
		"X-RapidAPI-Host": "api-football-v1.p.rapidapi.com"
	}

	response = requests.request("GET", url, headers=headers, params=querystring)

	resp_dict = json.loads(response.text)
	#print(json.dumps(resp_dict["response"],indent=4))
	teams = [(team["team"]["id"],team["team"]["name"]) for team in resp_dict["response"]]
	return teams

#get_teams_ids_for_league("39","2022")
```

```python
import time
def get_teams_without_draw_in_last_x_matchs(league,season,last):
	teams_list=get_teams_ids_for_league(league,season)

	url = "https://api-football-v1.p.rapidapi.com/v3/fixtures"
	teams_without_draw=[]
	for team in teams_list:

		querystring = {"league": league, "season":season, "team":team[0], "last":last
		}
		# query_by_team = {"team":"Arsenal","league": "39", "season":"2022"
		# }

		headers = {
			"X-RapidAPI-Key": "630322333bmsh894634ec4d1b679p17d2b3jsnf658c1fd7f38",
			"X-RapidAPI-Host": "api-football-v1.p.rapidapi.com"
		}

		response = requests.request("GET", url, headers=headers, params=querystring)
		resp_dict = json.loads(response.text)
		result=True
		for fixture in resp_dict["response"]:
			if fixture["score"]["fulltime"]["home"]==fixture["score"]["fulltime"]["away"]:
				result=False
				break
		if result==True:
			teams_without_draw.append(team[1])
		time.sleep(2)
	return teams_without_draw

get_teams_without_draw_in_last_x_matchs("39","2022","10")
```
